**Importing necessary libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Reading the data which is in csv form**

```python
df = pd.read_csv('/kaggle/input/amazon-sales-dataset/Amazon.csv')
df.head(5)
```

**Exploring the dataset shapes and datatypes.**

**Changing the data type of discount_percentage, rating, rating_count, actual_price, discounted_price.**

```python
# Converting discount_percentage to numerical data type
df['discount_percentage'] = df['discount_percentage'].str.rstrip('%').astype('float') / 100.0

# Converting rating to numerical data type
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Converting rating count to numerical data type
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Converting actual price to numerical data type
df['actual_price'] = pd.to_numeric(df['actual_price'], errors = 'coerce')

# Converting discount price to numerical data type
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors = 'coerce')

```

**Question 1: How does the discount percentage affect the rating of a product?**

```python
# Plot the relationship between discount_percentage and rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="discount_percentage", y="rating")
plt.title("Effect of Discount Percentage on Product Rating")
plt.xlabel("Discount Percentage")
plt.ylabel("Product Rating")
plt.grid(True)
plt.show()
```

**A scatter plot visualize the relationship between discount percentage and product rating. The plot showed no clear pattern indicating a direct relationship between higher discounts and higher ratings.**

**Question 4: What is the most common word in the positive and negative reviews?**

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import string
```

```python
# Define positive and negative reviews based on ratings
positive_reviews = df[df['rating'] >= 4]['review_content'].tolist()
negative_reviews = df[df['rating'] < 4]['review_content'].tolist()

# Combine all reviews into one text
positive_text = " ".join(positive_reviews)
negative_text = " ".join(negative_reviews)

# Tokenize the text
positive_tokens = word_tokenize(positive_text.lower())
negative_tokens = word_tokenize(negative_text.lower())

# Remove stopwords and punctuation
stop_words = set(stopwords.words('english'))
positive_words = [word for word in positive_tokens if word.isalpha() and word not in stop_words]
negative_words = [word for word in negative_tokens if word.isalpha() and word not in stop_words]

# Find the most common words
positive_word_counts = Counter(positive_words)
negative_word_counts = Counter(negative_words)

most_common_positive = positive_word_counts.most_common(1)
most_common_negative = negative_word_counts.most_common(1)

print(f"Most common word in positive reviews: {most_common_positive}")
print(f"Most common word in negative reviews: {most_common_negative}")
```

**Positive and negative reviews were separated based on rating thresholds. The most common word in positive reviews was "good", appearing 4751 times, and in negative reviews, it was also "good", appearing 1254 times.**

**Question 8: Is there a correlation between the length of a review and the rating given?**

```python
# Calculate the length of each review and store it in a new column
df['review_length'] = df['review_content'].apply(len)

# Compute the correlation between review length and rating
correlation = df['review_length'].corr(df['rating'])

print(f"Correlation between review length and rating: {correlation}")

# Visualize the correlation
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="review_length", y="rating")
plt.title("Correlation between Review Length and Rating")
plt.xlabel("Review Length")
plt.ylabel("Product Rating")
plt.grid(True)
plt.show()
```

**The correlation between review length and rating was calculated, resulting in a correlation coefficient of 0.077, indicating a weak positive correlation.**

**Question 9: Does having an image link affect the product rating?**

```python
# Step 1: Create a binary indicator for image link
df['has_image_link'] = df['img_link'].notna()

# Step 2: Analyze average ratings for products with and without image links
average_rating_with_image = df[df['has_image_link']]['rating'].mean()
average_rating_without_image = df[~df['has_image_link']]['rating'].mean()

print(f"Average rating for products with image link: {average_rating_with_image:.2f}")
print(f"Average rating for products without image link: {average_rating_without_image:.2f}")
```

**Products were analyzed based on whether they had an image link or not. Products with image links had an average rating of 4.10, while products without image links had NaN for average rating due to insufficient data.**
